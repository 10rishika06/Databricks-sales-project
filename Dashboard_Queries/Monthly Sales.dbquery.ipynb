```sql
-- Monthly Sales
SELECT 
    YEAR(TO_DATE(`ship date`, 'yyyy-MM-dd')) AS year,
    MONTH(TO_DATE(`ship date`, 'yyyy-MM-dd')) AS month,
    SUM(sales) AS monthly_sales
FROM sales_data
GROUP BY 
    YEAR(TO_DATE(`ship date`, 'yyyy-MM-dd')),
    MONTH(TO_DATE(`ship date`, 'yyyy-MM-dd'))
ORDER BY year, month;
```
